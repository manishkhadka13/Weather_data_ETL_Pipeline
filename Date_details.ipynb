```
import datetime
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType
from pyspark.pandas import date_range
from pyspark.sql.types import StructType,StructField,StringType,DoubleType,IntegerType,TimestampType,FloatType,ArrayType,DateType
```

```
date_schema=StructType([
    StructField('Date_id',IntegerType(),nullable=False),
    StructField('fulldate',DateType(),nullable=False),
    StructField('monthName',StringType(),nullable=False),
    StructField('month_num_of_year',IntegerType(),nullable=False),
    StructField('day_name_of_week',StringType(),nullable=False),
    StructField('day_num_of_week',IntegerType(),nullable=False),
    StructField('day_num_of_month',IntegerType(),nullable=False),
    StructField('day_num_of_year',IntegerType(),nullable=False),
    StructField('week_num_of_year',IntegerType(),nullable=False),
    StructField('quarterName',StringType(),nullable=False),
    StructField('calenderQuarter',IntegerType(),nullable=False),
    StructField('calenderYear',IntegerType(),nullable=False)
])
```

```
def load_date(start_date='2023-01-01',end_date='2023-12-30'):
    df_date = spark.createDataFrame(date_range(start=start_date, end=end_date).to_numpy(), ['Date'],date_schema)
    df_date = df_date.withColumn('Date_id', F.date_format('Date', 'yyyyMMdd').cast(IntegerType()))
    df_date = df_date.withColumn('fullDate', F.date_format('Date', 'yyyy-MM-dd').cast('DATE'))
    df_date = df_date.withColumn('monthName', F.date_format('Date', 'MMMM'))
    df_date = df_date.withColumn('month_num_of_year', F.month('Date'))
    df_date = df_date.withColumn('day_name_of_week', F.date_format('Date', 'EEEE'))
    df_date = df_date.withColumn('day_num_of_week', F.dayofweek('Date'))
    df_date = df_date.withColumn('day_num_of_month', F.dayofmonth('Date'))
    df_date = df_date.withColumn('day_num_of_year', F.dayofyear('Date'))
    df_date = df_date.withColumn('week_num_of_year', F.weekofyear('Date'))
    df_date = df_date.withColumn('quarterName', F.concat(F.lit('Q').cast('string'), F.quarter('Date')))
    df_date = df_date.withColumn('calenderQuarter', F.quarter('Date'))
    df_date = df_date.withColumn('calenderYear', F.year('Date'))
    df_date = df_date.drop('Date')
    
    df_date.write.format('delta').mode('overwrite').saveAsTable('Dim_date')
```

```
load_date()
```

```
%sql
select * from Dim_date;
```
