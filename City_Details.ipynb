```
from pyspark.sql.functions import col
from pyspark.sql.types import StructType,StructField,StringType,DoubleType,IntegerType,TimestampType,FloatType,ArrayType
```

####Fucntion to create dimension table for cities

```
city_schema=StructType([
    StructField("id",IntegerType(),nullable=False),
    StructField("country",StringType(),nullable=False),
    StructField("name",StringType(),nullable=False),
    StructField("latitudes",FloatType(),nullable=False),
    StructField("longitudes",FloatType(),nullable=False)
])
```

####Function to load the cities data into the dimension table

```
from pyspark.sql.functions import col
def cities_load(file_path,options):
    df=spark.createDataFrame([],city_schema)
    df=spark.read.format("json").options(**options).load(file_path).select('*','coord.*')
    df=df.select(col('id').cast('int'),col('country'),col('name'),col('lat').cast('float').alias('latitudes'),col('lon').cast('float').alias('longitude'))
    df=df.filter(df['country']=='NP')
    df.write.format('delta').mode('overwrite').saveAsTable('Dim_cities')
    
```

```
file_path="dbfs:/FileStore/city_list.json"
options={
  "multiline":True
}
cities_load(file_path,options)
```
