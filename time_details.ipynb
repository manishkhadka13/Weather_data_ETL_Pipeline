```
from pyspark.sql.types import StructType,StructField,StringType,DoubleType,IntegerType,TimestampType,FloatType,ArrayType
from pyspark.sql.functions import col
```

```
time_schema=StructType([
    StructField('time_id',IntegerType(),nullable=False),
    StructField('start_time',StringType(),nullable=False),
    StructField('end_time',StringType(),nullable=False)
])
```

```
def load_time():
    time_list = []
    for hour in range(24):
        start_time = f"{hour:02d}:00"  # Format start time as HH:00
        end_time = f"{hour:02d}:59"  # Format end time as HH:59
        time_list.append({"time_id": hour, "start_time": start_time, "end_time": end_time})
    df = spark.createDataFrame(time_list,time_schema)
    df = df.withColumn('time_id', col('time_id').cast('int'))
    df.write.format('delta').mode('overwrite').saveAsTable('Dim_time')
```

```
load_time()
```

```
%sql
select * from Dim_time;
```
